# Setup the enviornment by importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Read the dataseet into a DataFrame and start exploring the data

```python
recent_grads = pd.read_csv("recent-grads.csv")

recent_grads.iloc[0, :]

print(recent_grads.head())

print(recent_grads.tail())
```

### Generate summary statistics for all of the numeric columns

```python
print(recent_grads.describe())
```

# Drop rows with missing values

```python
raw_data_count = recent_grads.shape[0]
print("Number of rows in raw dataframe: ", raw_data_count) # orignially read 173

# use dataframe.dropna() to drop rows containing missing values and assign back to original dataframe
recent_grads = recent_grads.dropna()

cleaned_data_count = recent_grads.shape[0]
print("Number of rows in cleaned dataframe: ", cleaned_data_count)
```

# Simple plots

```python
# These will be plotted individually due to the `%matplotlib inline` magic in the first cell
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')

recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))

ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```
